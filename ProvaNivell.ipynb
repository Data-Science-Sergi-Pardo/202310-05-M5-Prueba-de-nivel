```python
import pandas as pd
from sklearn.datasets import load_diabetes
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataSet = load_diabetes()
data = dataSet.data
dataNames = dataSet.feature_names
```

```python
df = pd.DataFrame(data=data, columns=dataNames)
df.head(10)
```

```python
df.info()
```

```python
def sex_unitari(x):
    if x < 0:
        resultat = 0
    else:
        resultat = 1
    return resultat

df['sex'] = df['sex'].apply(sex_unitari)
```

```python
df.head(6)
```

```python
df.corr()
```

```python
df.isnull().sum()
```

```python
correlacion = round(df.corr(),2)

sns.heatmap(correlacion, annot=True)
```

```python
fig = sns.kdeplot(df['s2'], shade=True, color="r")
fig = sns.kdeplot(df['s4'], shade=True, color="b")
plt.show()
```

```python
sns.boxplot(y=df["bmi"] )
```
